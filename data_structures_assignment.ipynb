Task 3 : Product Pricing (Dictionaries)

Created a dictionary showing prices of products

```python
price_dict={"Laptop":40000, "T-Shirt":500, "Phone":6000, "Chair":2500, "Jeans":1000, "Smartphone":18000}
type(price_dict)
```

add a new product with price

```python
price_dict["Phone Case"]=200
print(price_dict)
```

update price of an existing product

```python
price_dict["Jeans"]=1200
print(price_dict)
```

remove a product and handle the case if the product does not exist

```python
product_to_be_removed= "Phone"
if product_to_be_removed in price_dict:
    del product_to_be_removed
    print("Product deleted from the dictionary successfully")
else: 
    print("Product not found in the dictionary")
```

```python
product_to_be_removed= "Blanket"
if product_to_be_removed in price_dict:
    del product_to_be_removed
    print("Product deleted from the dictionary successfully")
else: 
    print("Product not found in the dictionary")
```

print average price of all products 

```python
total_price=sum(price_dict.values())
count_products=len(price_dict.items())
avg_price=total_price/count_products
print("Average price of products is",avg_price)
```

print product with maximum and minimum price

```python
max_product=max(price_dict.values())
print("Maximum product price is ",max_product)
min_product=min(price_dict.values())
print("Minimum product price is ",min_product)
```

Create catalog list of tuples (product_name, price, category)

```python
catalog=[]
for product_name in products:
    if(product_name=="Laptop"):
        catalog.append((product_name,price_dict.get(product_name),"Electronics"))
    elif(product_name=="T-Shirt"):
        catalog.append((product_name,price_dict.get(product_name),"Clothing"))
    elif(product_name=="Phone"):
        catalog.append((product_name,price_dict.get(product_name),"Electronics"))
    elif(product_name=="Chair"):
        catalog.append((product_name,price_dict.get(product_name),"Furniture"))
    elif(product_name=="Jeans"):
        catalog.append((product_name,price_dict.get(product_name),"Clothing"))
    elif(product_name=="Smartphone"):
        catalog.append((product_name,price_dict.get(product_name),"Electronics"))
    elif(product_name=="Phone Case"):
        catalog.append((product_name,price_dict.get(product_name),"Others"))
```

```python
catalog
```

Create category_to_products dictionary


```python
category_to_products={}

for product_name, price, category in catalog:
    if category not in category_to_products:
        category_to_products[category]=[]
    category_to_products[category].append(product_name)   
```

print all products of the category that has maximum products

```python
max_category = max(category_to_products, key=lambda k: len(category_to_products[k]))

print("Category with maximum products:", max_category)
print("Products in that category:")

for product in category_to_products[max_category]:
    print(product)
```
