Task 1:Discount Rule(if/elif/else)

```python
order_amount=int(input("Please enter order amount"))
```

```python
print(order_amount)
```

Apply specified discount rules

```python
if order_amount >= 2000:
    print("15% discount")
elif order_amount >= 1500 and order_amount <2000:
    print("10% discount")
elif order_amount >= 1000 and order_amount <1500:
    print("7% discount")
else:
    print("0% discount")
```

Convert the input to a number and handle the case displaying error when the input is not number

```python
while True:
    order_amount = input("Please enter order amount: ")
    
    if order_amount.isdigit():
        order_amount = int(order_amount)
        if order_amount >= 2000:
            print("15% discount")
        elif order_amount >= 1500 and order_amount <2000:
            print("10% discount")
        elif order_amount >= 1000 and order_amount <1500:
            print("7% discount")
        else:
            print("0% discount")
        break
    else:
        print("Input should be a number")

```

Task 2 : Process multiple orders

Apply discount from task 1 to the list of orders and print summary table

```python
orders = [1200,2500,800,1750,3000]

print("order_amount -> discount -> final_amount")
for order_amount in orders:
    if order_amount >= 2000:
        print(f"{order_amount} -> 15% discount -> {order_amount-(order_amount*0.15)}")
    elif order_amount >= 1500 and order_amount <2000:
        print(f"{order_amount} -> 10% discount -> {order_amount-(order_amount*0.1)}")
    elif order_amount >= 1000 and order_amount <1500:
        print(f"{order_amount} -> 7% discount -> {order_amount-(order_amount*0.07)}")
    else:
        print(f"{order_amount} -> 0% discount -> {order_amount}")
```

Also compute and print total revenue after discounts

```python
orders = [1200,2500,800,1750,3000]
total_revenue=0

print("order_amount -> discount -> final_amount")
for order_amount in orders:
    if order_amount >= 2000:
        total_revenue+=order_amount-(order_amount*0.15)
        print(f"{order_amount} -> 15% discount -> {total_revenue}")
    elif order_amount >= 1500 and order_amount <2000:
        total_revenue+=order_amount-(order_amount*0.1)
        print(f"{order_amount} -> 10% discount -> {total_revenue}")
    elif order_amount >= 1000 and order_amount <1500:
        total_revenue+=order_amount-(order_amount*0.07)
        print(f"{order_amount} -> 7% discount -> {total_revenue}")
    else:
        total_revenue+=order_amount
        print(f"{order_amount} -> 0% discount -> {total_revenue}")
print("Total revenue after discounts is ",total_revenue)
```

Task 3 : User Menu

```python
orders=[]
print("Menu options\n1 - Add new order amount\n2 - Show all orders and totals after applying discount\nq - Quit\n")
    
while True:
    menu_option = str(input("Choose an option to proceed"))
    
    if(menu_option=="1"):
        order_amount=int(input("Please enter new order amount"))
        orders.append(order_amount)
        print("Order added succefully")
        print("Updated orders ",orders,"\n")
        continue
    elif(menu_option=="2"):
        total_revenue=0

        print("order_amount -> discount -> final_amount")
        for order_amount in orders:
            if order_amount >= 2000:
                total_revenue+=order_amount-(order_amount*0.15)
                print(f"{order_amount} -> 15% discount -> {total_revenue}")
            elif order_amount >= 1500 and order_amount <2000:
                total_revenue+=order_amount-(order_amount*0.1)
                print(f"{order_amount} -> 10% discount -> {total_revenue}")
            elif order_amount >= 1000 and order_amount <1500:
                total_revenue+=order_amount-(order_amount*0.07)
                print(f"{order_amount} -> 7% discount -> {total_revenue}")
            else:
                total_revenue+=order_amount
                print(f"{order_amount} -> 0% discount -> {total_revenue}")
        print("Total revenue after discounts is ",total_revenue,"\n")
        continue
    elif(menu_option=="q"):
        break


```

Task 4 : Loop control with conditions

```python
daily=[200,150,0,400,50,-1,300]
total_sales=0

for sales in daily:
    if sales == -1:
        print("Corrupted data!")
        break
    elif sales == 0:
        print("No sales")
        continue
    else:
        total_sales+=sales
        print("Updated sales ",total_sales)

```
