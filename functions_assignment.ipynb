Task 1: Basic Function: Price After Discount

```python
def apply_discount(price,discount_price=5):
    if(discount_price>60):
        discount_price=60
    price_after_discount= price - (price*discount_price)/100
    return price_after_discount


```

```python
apply_discount(1500,10)
```

Task 2: Recursive Function: Factorial Utility

```python
def factorial(n):
    if(n<0):
        print("Factorial is undefined for negative numbers")
    elif(n==0 or n==1):
        return 1
    else: 
        return n * factorial(n-1)
```

```python
factorial(5)
```

Task 3: Lambda Function: GST Calculator

```python
gst = lambda price: price + (0.18 * price)

print(gst(100))
```

```python
final_price = lambda price,discount:price + (price * 0.18) - (price * discount * 0.01)

print(final_price(100, 18))
```

Task 4 : using Map() apply GST to list of prices

```python
prices = [100,250,400,1200,50]

prices_with_gst = list(map(gst,prices))
print("Original Prices are ",prices)
print("Prices after GST are ",prices_with_gst)
```

Task 5: using Filter():Filter expensive products

```python
prices = [100,250,400,1200,50,2000,850]

products_greater_than_500 = list(filter(lambda p:p>500,prices))
products_less_than_500 = list(filter(lambda p:p<=500,prices))
print("Product price greater than 500 ",products_greater_than_500)
print("Product price less than 500 ",products_less_than_500)
```

Task 6 : Combined Utlity function

```python
def process_prices(prices):
    discounted_prices = list(map(lambda x:x-x*0.01,prices))
    filtered_prices = list(filter(lambda x:x>300,discounted_prices))
    print("Discounted prices are",discounted_prices)
    print("Prices above 300 are",filtered_prices)
```

```python
process_prices([100,500,900,50,750])
```

Task 7: Mini Problem: Menu using Functions

```python
def add_price(prices_list,prices):
    prices_list.append(prices)
    return prices_list

def get_average_price(prices_list):
    sum=0
    for p in prices_list:
        sum=sum+p
    return sum/len(prices_list)

def get_max_price(prices_list):
    return max(prices_list)
```

```python
print("Menu options\n1 - Add price\n2 - Show average price\n3 - Show highest price\nq - Quit\n")
prices_list=[]    
while True:
    menu_option = str(input("Choose from the menu"))
    
    if(menu_option=="1"):
        price = int(input("Enter new price"))
        add_price(prices_list,price)
        print("Price added succefully")
        print("Updated price list ",prices_list,"\n")
        continue
    elif(menu_option=="2"):
        print("Average price is ",get_average_price(prices_list),"\n")
        continue
    elif(menu_option=="3"):
        print("Highest price is ",get_max_price(prices_list),"\n")
        continue
    elif(menu_option=="q"):
        break
```
